```python
import pandas as pd
import random

# Listes de valeurs possibles
intitule_poste = ["Data Ingenieur", "Data Engineer", "Data Analyst", "Data Analyste", "Data Scientist"]
departements = ["Ain", "Aisne", "Allier", "Alpes-de-Haute-Provence", "Hauts-de-Seine", "Seine-Saint-Denis", "Val-de-Marne", "Val-d'Oise"]
niveau_experience = ["Junior", "Intermédiaire", "Senior"]
secteur_d_activite = ["Tech", "Finance", "Industrie", "Santé", "Marketing", "Autre"]
taille_entreprise = ["Startup", "PME", "Grande entreprise"]
type_de_contrat = ["CDI", "CDD", "Freelance", "Stage", "Alternance"]
teletravail = ["Complet", "Partiel", "Occasionnel", "Aucun"]

def generer_nom_entreprise():
    prefixes = ["Tech", "Data", "AI", "Cloud", "Innov", "Cyber", "Smart"]
    suffixes = ["Corp", "Solutions", "Industries", "Labs", "Consulting", "Systems", "Services"]
    return random.choice(prefixes) + random.choice(suffixes) + " " + str(random.randint(1, 99))

def generer_description(row):
    return (f"📌 **Poste** : {row['intitule_poste']} ({row['niveau_experience']})\n"
            f"🏢 **Entreprise** : {row['nom_entreprise']} ({row['taille_entreprise']})\n"
            f"📍 **Localisation** : {row['departement']}\n"
            f"📊 **Secteur** : {row['secteur_d_activite']}\n"
            f"📝 **Type de contrat** : {row['type_de_contrat']}\n"
            f"💰 **Salaire estimé** : {row['salaire']} €\n"
            f"💻 **Télétravail** : {row['teletravail']}\n\n"
            f"🔍 **Description** : Nous recherchons un {row['intitule_poste']} motivé pour rejoindre notre équipe dynamique "
            f"chez {row['nom_entreprise']}. Vous serez amené(e) à travailler sur des projets innovants dans le domaine "
            f"{row['secteur_d_activite'].lower()}. Vous aurez l'opportunité de collaborer avec des experts et "
            f"de développer vos compétences dans un environnement stimulant. \n\n"
            f"🎯 **Missions** : \n"
            f"- Analyser et interpréter les données pour aider à la prise de décision\n"
            f"- Développer des modèles prédictifs et des visualisations avancées\n"
            f"- Collaborer avec les équipes métiers pour comprendre leurs besoins\n"
            f"- Automatiser et optimiser les processus analytiques\n\n"
            f"✅ **Profil recherché** : \n"
            f"- Expérience : {row['niveau_experience']} en data science ou analyse de données\n"
            f"- Compétences : Python, SQL, Tableau, Machine Learning\n"
            f"- Esprit d’analyse, rigueur et bon relationnel\n\n"
            f"📩 **Postulez dès maintenant !**")

# Générer 200 lignes de données fictives
data = {
    "intitule_poste": [random.choice(intitule_poste) for _ in range(200)],
    "departement": [random.choice(departements) for _ in range(200)],
    "niveau_experience": [random.choice(niveau_experience) for _ in range(200)],
    "secteur_d_activite": [random.choice(secteur_d_activite) for _ in range(200)],
    "taille_entreprise": [random.choice(taille_entreprise) for _ in range(200)],
    "type_de_contrat": [random.choice(type_de_contrat) for _ in range(200)],
    "teletravail": [random.choice(teletravail) for _ in range(200)],
    "salaire": [random.randint(30000, 100000) for _ in range(200)],
    "nom_entreprise": [generer_nom_entreprise() for _ in range(200)]
}

# Création du DataFrame
df = pd.DataFrame(data)

# Ajout des colonnes annonce et lien
df["annonce"] = df.apply(generer_description, axis=1)
df["lien"] = ["https://job-platform.com/offre-" + str(i) for i in range(1, 201)]

# Affichage des premières lignes
df.head()
```
